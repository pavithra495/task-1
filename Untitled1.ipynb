```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
model_name = "gpt2"  # You can also try "gpt2-medium" or others

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# To avoid warnings
tokenizer.pad_token = tokenizer.eos_token
model.resize_token_embeddings(len(tokenizer))
```

```python
from datasets import Dataset

def load_text_dataset(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()
    # Create a dataset from the text
    dataset = Dataset.from_dict({"text": [text]})
    return dataset

train_dataset = load_text_dataset("internship (1).txt")

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=128)

tokenized_dataset = train_dataset.map(tokenize_function, batched=True, remove_columns=["text"])

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False,
)
```

Now that the file path is corrected, you can run the training cell again to fine-tune the GPT-2 model on your custom dataset.

```python
training_args = TrainingArguments(
    output_dir="./gpt2-finetuned",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=500,
    save_total_limit=2,
    logging_dir='./logs',
    report_to="none",  # Disable reporting to wandb
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_dataset,  # Use the tokenized dataset
)

trainer.train()
```

```python
trainer.save_model("./gpt2-finetuned")
tokenizer.save_pretrained("./gpt2-finetuned")
```

```python
from transformers import pipeline

generator = pipeline('text-generation', model="./gpt2-finetuned", tokenizer="./gpt2-finetuned")
prompt = "Once upon a time"
results = generator(prompt, max_length=100, num_return_sequences=1)

print(results[0]['generated_text'])
```
